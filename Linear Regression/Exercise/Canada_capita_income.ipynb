```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head(5)
```

```python
df.rename(columns={'per capita income (US$)' : 'per_capita_income'}, inplace=True)
```

```python
plt.xlabel('Year')
plt.ylabel('Capita income ($)')
plt.scatter(df.year, df.per_capita_income,color='green',marker='+')
```

```python
df1 = df.drop('per_capita_income',axis=1)
df1.head()
```

```python
income = df.per_capita_income
income.head(5)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df1,income)
```

```python
reg.predict([[2020]])
```

```python
reg.predict([[2023]])
```
